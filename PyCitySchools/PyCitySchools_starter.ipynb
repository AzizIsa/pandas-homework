```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools

total_schools = len(school_data_complete["school_name"].unique())
#Calculate the total number of students
total_students = school_data_complete["student_name"].count()

#Calculate the total budget
total_budget = school_data_complete["budget"].unique().sum()

# Calculate the average math score
avr_math_score = school_data_complete["math_score"].mean()

# Calculate the average reading score
avr_reading_score = school_data_complete["reading_score"].mean()

#Calculate the percentage of students with a passing math score (70 or greater)
avr__math_count =school_data_complete.loc[school_data_complete["math_score"]>=70]["math_score"].count()
math_percent_pass = (avr__math_count/total_students)*100

#Calculate the percentage of students with a passing reading score (70 or greater)
avr__read_pass_count =school_data_complete.loc[school_data_complete["reading_score"]>=70]["reading_score"].count()
read_percent_pass =(avr__read_pass_count/total_students)*100

#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_pass=(math_percent_pass+read_percent_pass)/2
dis_summary = pd.DataFrame({"":[0],"Total schools":total_schools, "Total Students":total_students,"Total Budget":total_budget,"Average Math Score":avr_math_score,"Average Reading Score":avr_reading_score,"% Passing Math":math_percent_pass,"% Passing Reading":read_percent_pass,"Overall Passing":overall_pass})
dis_summary.set_index("").style.format({"Total Budget": "${:,.2f}","Total Students":"{:,}"})
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
# School Name
school_group = school_data_complete.groupby("school_name")
#School Type
school_type = school_group["type"].unique()
#Total Students
school_size = school_group["size"].mean()
#Total School Budget
school_budget = school_group["budget"].mean()
#Per Student Budget
per_student_budget = school_budget/school_size

#Average Math Score
avr_math_score = (school_group["math_score"].sum())/ school_size
#Average Reading Score
avr_read_score = (school_group["reading_score"].sum())/school_size

#% Passing Math
math_pass_perschool = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name").count()["student_name"]
percent_math_pass = (math_pass_perschool/school_size) * 100
#% Passing Reading
read_pass_perschool = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name").count()["student_name"]
percent_read_pass = (read_pass_perschool/school_size) *100

#Overall Passing Rate (Average of the above two)
percent_overall_pass = (percent_math_pass+percent_read_pass)/2

#Create a dataframe to hold the above results
summary_table =pd.DataFrame({"School Type" : school_type, "Total Students": school_size, "Total School Budget":school_budget, "Per Student Budget": per_student_budget,
                 "Average Math Score": avr_math_score, "Average Reading Score": avr_read_score, "% Passing Math":percent_read_pass , "%Passing Reading":percent_read_pass , "% Overall Passing Rate": percent_overall_pass})

# Sort the results to show the top performing schools
top_performing_schools = summary_table.sort_values("% Overall Passing Rate", ascending = False)
top_performing_schools.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
# Sort the results to show the Bottom performing schools
bottom_performing_schools = summary_table.sort_values("% Overall Passing Rate", ascending = True)
bottom_performing_schools.head()
```
